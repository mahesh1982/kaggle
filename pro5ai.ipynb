```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stop = stopwords.words("english")
stop.extend(['from', 'subject', 're', 'edu', 'use'])
stop = set(stop)
```

```python
def clean_text(file_path_name, column):
    # reading a file line by line
    with open(file_path_name) as file_in:
        for line in file_in:
            # convert line to lower case
            line = line.lower()
            # Remove all punctuation
            line = re.sub(r'[^\w\s]','',line)
            # Tokenization, if nothing is specified, then it will be splitted as per space
            line_1 = nltk.word_tokenize(line)
            print (line_1)
            filtered_words = [word.lower() for word in line_1 if word not in stop]
ret_lst = clean_text(r"/kaggle/input/spam-emails/spam.csv", "Message")
print (ret_lst)
```
